```python
from pycaret.classification import *
```

```python
from pycaret.datasets import get_data
```

```python
data=get_data('poker')
```

```python
s=ClassificationExperiment()
```

```python
s.setup(data, target='CLASS', session_id=123)
```

```python
model= s.compare_models()
```

```python
s.evaluate_model(model)
```

```python
s.save_model(model, 'poker_classification_model')
```
